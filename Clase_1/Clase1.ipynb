```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función de la señal triangular
def signal_func(t):
    return np.mod(2*t, 2)  # señal triangular

# Definir el rango de tiempo
t = np.linspace(0, 1, 30) # desde 0 a 1 segundo con 30 puntos

# Generar la señal
signal = signal_func(t)

# Graficar la señal
plt.plot(t, signal)
plt.stem(t, signal)
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Señal analógica triangular')
plt.grid(True)
plt.show()
```
